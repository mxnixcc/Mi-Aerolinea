```python
#Números primos
def es_primo(n):
  if n < 2:
    return False
  for i in range(2, int(n**0.5)+1):
    if n % i == 0:
      return False
  return True
```

```python
es_primo(1949)
```

Sec de Fibobacci

```python
def fibonacci(n):
  fib_seq = [0,1]
  for i in range(2, n):
    fib_seq.append(fib_seq[-1] + fib_seq[-2])
  return fib_seq
```

```python
fibonacci(15)
```

```python
#Contar palabras de un texto
def contar_palabras(texto):
  palabras = texto.split()
  conteo = {}
  for palabra in palabras:
    conteo[palabra] = conteo.get(palabra, 0) + 1
  return conteo
```

```python
texto = "Hola Juan Pablo Montecino Hola Pablo"
contar_palabras(texto)
```

```python
#Palindromo
def es_palindromo(palabra):
  return palabra == palabra[::-1]
```

```python
#Factorial
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
#Suma de los digitos
def suma_digitos(n):
  return sum(int(digito) for digito in str(n))
```

```python
suma_digitos(1234)
```

```python
#Encontrar el mayor de tres números
def mayor_de_tres(a, b, c):
  return max(a, b, c)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights.dropna(subset=['arr_delay', 'carrier', 'dest', 'tailnum'], inplace = True)
```

```python
flights_airlines = flights.merge(airlines, on = 'carrier')
flights_airports = flights.merge(airports, left_on='dest', right_on='faa')
flights_planes = flights.merge(planes, on='tailnum', how='left')
flights_weather = flights.merge(weather, on= ['year', 'month', 'day','hour', 'origin'], how='left')
```

```python
punctuality = flights_airlines.groupby('name')['arr_delay'].mean().sort_values()
punctuality = punctuality.head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=punctuality.values, y =punctuality.index, palette='Greens_d')
plt.title('Top 5 Aerolineas con menor retraso promedio')
plt.xlabel('Retraso promedio (minutos)')
plt.ylabel('Aerolinea')
plt.show()
```

```python
flights_airports['retrasado'] = flights_airports['arr_delay'] > 15
retrasos = flights_airports.groupby('name')['retrasado'].sum().sort_values(ascending = False).head(5)
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x=retrasos.values, y =retrasos.index, palette='Reds_d')
plt.title('Top 5 Aeropuertos con mas vuelos retrasados')
plt.xlabel('Cantidad de vuelos retrasados')
plt.ylabel('Aeropuerto')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=flights_weather, x='temp', y ='arr_delay', alpha=0.2)
plt.title('Retrasos según temperatura')
plt.xlabel('Temperatura (F°)')
plt.ylabel('Retraso en llegada (minutos)')
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(data=flights_weather, x='wind_speed', y ='arr_delay', alpha=0.2)
plt.title('Retrasos según velocidad del viento')
plt.xlabel('Velocidad del viento (mph)')
plt.ylabel('Retraso en llegada (minutos)')
plt.show()
```

```python
modelos= flights_planes.groupby('model')['arr_delay'].mean().dropna()
modelos_top = modelos.sort_values(ascending = False).head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = modelos_top.values, y = modelos_top.index, palette='Purples_d')
plt.title('Top 5 Aviones con mayor retraso pomedio')
plt.xlabel('Retraso promedio (minutos)')
plt.ylabel('Modelo Avion')
plt.show()
```
